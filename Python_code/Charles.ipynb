```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
```

```python
# data load

df = pd.read_csv('C:/Users/charl/Documents/GitHub/Workshop3_Dylan/data/steam.csv', index_col=0)
df
```

```python
# data processing
df["positive_ratio"] = df["positive_ratings"] / (df["positive_ratings"] + df["negative_ratings"])
```

```python
df
```

```python
df = df[(df['positive_ratings'] + df['negative_ratings']) >= 500]
df
```

```python
df_encoded = pd.get_dummies(df)
df_encoded
```

```python
X = df.drop(columns=["positive_ratio"], axis=1)
y = df["positive_ratio"]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(X.shape[1], input_dim=X.shape[1], activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, )
])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
X = X.apply(label_encoder.fit_transform)
X


```

```python
y = y.astype(float)
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
model.fit(X, y, epochs=100, batch_size=256, validation_split=0.2)
```
