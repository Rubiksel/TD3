```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.read_csv("steam.csv", sep = ",")
display(df.head())
```

```python
df.columns
```

```python
df["positive_ratio"] = df["positive_ratings"] / (df["positive_ratings"] + df["negative_ratings"])
display(df)
```

```python
df = df[(df["positive_ratings"] + df["negative_ratings"]) >= 500]
display(df)
```

```python
df_encoded = pd.get_dummies(df)
display(df_encoded)
```

```python
X = df.drop(columns = ["positive_ratio"], axis = 1)
y = df["positive_ratio"]
```

```python
label_encoder = LabelEncoder()
X = X.apply(label_encoder.fit_transform)
display(X)
```

```python
y = y.astype(float)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 6)

rfc = RandomForestRegressor(n_estimators=100, max_depth = 25)

rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
results_df = pd.DataFrame({
    "True Value": y_test,
    "Predicted Value": y_pred,
    "Name": X_test['name'],
    "Delta Error": y_test-y_pred
})

display(results_df)
```
